```python
#defining locations of csv file with correct patient ids and folder where your images are, also define where you want the removed folders
csv_path = r"C:\Users\Diederik\OneDrive\Bureaublad\studie tn\Minor vakken Porto\IA CAD\GITHUB\LAB\id_patients_filtered.csv"
folder_path = r"C:\Users\Diederik\OneDrive\Bureaublad\studie tn\Minor vakken Porto\IA CAD\Images\manifest-1600709154662\LIDC-IDRI"
destination_folder = r"C:\Users\Diederik\OneDrive\Bureaublad\studie tn\Minor vakken Porto\IA CAD\Removed images"


def compare_to_delete(csv_path, folder_path,  destination_folder):
    
    #converting csv and file list into set()
    csv_ids = set(np.genfromtxt(csv_path, delimiter=',', dtype='str'))
    folder_ids = set(folder_name for folder_name in os.listdir(folder_path))
    ids_to_move = folder_ids - csv_ids

    #loop for moving folders     
    for folder_ids in ids_to_move:
            folder_to_move = os.path.join(folder_path, folder_ids)
            destination_path = os.path.join(destination_folder, folder_ids)
            try:
                shutil.move(folder_to_move, destination_path)
                print(f"Moved folder '{folder_ids}' to {destination_folder}")
            except Exception as e:
                print(f"Error moving folder '{folder_ids}': {e}")
```

```python
compare_to_delete(csv_path, folder_path, destination_folder)
```
